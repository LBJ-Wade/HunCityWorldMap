```python
import numpy as np
from matplotlib import pyplot as plt
import healpy as hp
```

```python
city_names = []
lats = []
longs = []
with open("hun_cities.txt", 'r') as ff:
    for line in ff:
        ll = line.split('\t')
        city_names.append(ll[1])
        lats.append(float(ll[3]))
        longs.append(float(ll[4]))
```

```python
print(city_names)
print(lats)
print(longs)
```

```python
NSIDE=8
Npix=hp.nside2npix(NSIDE)

m=np.zeros(Npix)

hp.mollview(m)
for lon,lat, name in zip(longs, lats, city_names):
    hp.projscatter([lon, lat], lonlat=True, marker='x', label=name)

plt.gca().legend()
```
